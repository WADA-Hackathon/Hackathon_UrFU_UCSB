**Загрузка и соединение данных**

```python
import pandas as pd
```

Первая таблица

```python
df_attack = pd.read_excel('/content/SWaT_Dataset_Attack_v0.xlsx', header=0)
df_attack.head(5)
```

```python
# add header
new_header = df_attack.iloc[0]
df_attack = df_attack[1:]
df_attack.columns = new_header
```

Вторая таблица

```python
df_normal_1 = pd.read_excel('/content/SWaT_Dataset_Normal_v0.xlsx')
df_normal_1.head(5)
```

```python
# add header
new_header = df_normal_1.iloc[0]
df_normal_1 = df_normal_1[1:]
df_normal_1.columns = new_header
```

Третья таблица

```python
df_normal_2 = pd.read_excel('/content/SWaT_Dataset_Normal_v1.xlsx')
df_normal_2.head(5)
```
